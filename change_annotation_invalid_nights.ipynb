```python
import pandas as pd
```

```python

data_cleaning_df = pd.read_csv("all_data_cleaning.csv")
sleeplog_df = pd.read_csv('sleep_logs_with_dates.csv')

for index, row in data_cleaning_df.iterrows():
    identifier = row["ID"]
    column_index = row[1:] 
    column_index = column_index.dropna()
    tam = len(column_index)
    
    if identifier in sleeplog_df["ID"].values:
        #print(identifier)
        for ii in range(tam):
            ind = int((column_index[ii] * 2)-1)
            sleep_content = sleeplog_df[sleeplog_df["ID"] == identifier]

            prev_sleep = sleep_content.iloc[0, ind]
            #print(prev_sleep)
            try:
                prev_date = str(prev_sleep[0:10])
                new_datetime = prev_date + 'T03:00:00'

                sleeplog_df.loc[sleeplog_df["ID"] == identifier, sleeplog_df.columns[ind]] = new_datetime
                sleeplog_df.loc[sleeplog_df["ID"] == identifier, sleeplog_df.columns[ind+1]] = new_datetime
            except:
                print(f"Invalid date format for identifier {identifier}, column index {ind}")
           
    sleeplog_df.to_csv("modified_sleeplog.csv", index=False)

```
