```python
import numpy as np
from datetime import datetime, timedelta
import csv
```

```python
## important: need to remove the header from the 'new_sleeplog.csv' file
```

```python

to_datetime = lambda t: datetime.strptime(t, '%Y-%m-%dT%H:%M:%S')
add_day = lambda t: t + timedelta(days=1)
is_possible = lambda t1, t2: bool((t2 - t1).total_seconds() >= 0)

fl = 'new_sleeplog.csv'

logs = {}
with open(fl) as fhandle:
    for ln in fhandle.readlines():
        tmpln = ln.strip().split(',')
        times = []
        # Doing this in a loop instead of comprehension so we can exit when
        # we hit our first broken timestamp (erroneous inclusions)
        for t in tmpln[1:]: 
            try:
                times += [to_datetime(t)]
            except ValueError as e:
                break
        logs[tmpln[0]] = times


def update_dates(log, fixes=0):
    new_log = [log[0]]
    diffs = [is_possible(t1, t2) for t1, t2 in zip(log, log[1:])]
    for idx, d in enumerate(diffs):
        if not d:
            fixes += 1
            new_log += [add_day(log[idx+1])]
        else:
            new_log += [log[idx+1]]

    diffs2 = [is_possible(t1, t2) for t1, t2 in zip(new_log, new_log[1:])]
    if not all(diffs2):
        new_log, fixes = update_dates(new_log, fixes=fixes)
    return new_log, fixes


corrected_logs = {}
for sub, log in logs.items():
    cl, fixes = update_dates(log, fixes=0)
    corrected_logs[sub] = [c.strftime('%Y-%m-%dT%H:%M:%S') for c in cl]

    print(fixes)
    print(sub, fixes, corrected_logs[sub])
    print(all([is_possible(t1, t2) for t1, t2 in zip(cl, cl[1:])]))

with open('sleep_logs_with_dates.csv', 'w') as fhandle:
    writer = csv.writer(fhandle)
    for key, value in corrected_logs.items():
       writer.writerow([key, value])
```
